```python
def bubbleSort(arr):
    n = len(arr)

    for i in range(n-1):
        swapped = False

        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                # Tukar elemen
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        # Jika tidak ada dua elemen yang ditukar oleh loop dalam, array sudah diurutkan
        if not swapped:
            break

# Kode driver untuk diuji di atas
arr = [64, 34, 25, 12, 22, 11, 90]
bubbleSort(arr)
print("Sorted array is:")
for i in range(len(arr)):
    print("%d" % arr[i], end=" ")
```

```python
def isSorted(arr):
    # Periksa apakah array sudah diurutkan
    return all(arr[i] <= arr[i + 1] for i in range(len(arr) - 1))

def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[min_idx] > arr[j]:
                min_idx = j

        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Contoh penggunaan:
A = [64, 25, 12, 22, 11]

# Periksa apakah array sudah diurutkan sebelum menyortir
if not isSorted(A):
    selectionSort(A)

# Print array yang diurutkan
print("Sorted array:")
for num in A:
    print("%d" % num, end=" ")
```

```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key


arr2 = [45, 23, 78, 12, 56]
insertionSort(arr2)

print("Sorted array:")
for num in arr2:
    print("%d" % num)
```

Bubble Sort:

Langkah 1: Bandingkan elemen pertama dan elemen kedua. Jika elemen pertama lebih besar dari elemen kedua, tukar mereka.
Langkah 2: Lanjutkan untuk membandingkan dan menukar elemen berdekatan hingga elemen terakhir.
Langkah 3: Ulangi langkah-langkah 1 dan 2 untuk setiap pasangan elemen dalam array hingga seluruh array diurutkan.
Proses ini mirip dengan gelembung yang naik ke permukaan, sehingga namanya Bubble Sort.

Selection Sort:

Langkah 1: Temukan elemen terkecil dalam array dan tukar dengan elemen pertama.
Langkah 2: Temukan elemen terkecil di sisa array (setelah elemen pertama) dan tukar dengan elemen kedua.
Langkah 3: Teruskan proses ini, menemukan elemen terkecil di sisa array dan menukarnya dengan elemen berikutnya hingga seluruh array diurutkan.
Pada setiap langkah, Selection Sort memilih elemen terkecil dan menempatkannya di posisi yang sesuai dalam array.

Insertion Sort:

Langkah 1: Bagi array menjadi dua bagian: bagian terurut dan bagian belum terurut.
Langkah 2: Ambil elemen pertama dari bagian belum terurut dan sisipkan ke tempat yang tepat dalam bagian terurut.
Langkah 3: Ulangi langkah 2 untuk setiap elemen dalam bagian belum terurut hingga seluruh array diurutkan.
